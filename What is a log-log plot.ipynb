A Log-Log plot is a graph that represents a relationship between two variables once a logarithm has been applied to them. 
First make a plot of two variables with simple linear relationship. 

Start with $y=mx+b$

Second, a parabola: $y=m_{2}x^{2} + m_{1}x +b_{0}$

```python
import numpy as np
# Set up matplotlib and use a nicer set of plot parameters
import matplotlib
#matplotlib.rc_file("../../templates/matplotlibrc")
import matplotlib.pyplot as plt
matplotlib.rc('text', usetex=True)
```

```python
# declare some variables
m=1.1
b=-2
# make 'arrays' of the data. This is really an x table and a y table
# x | y
# 0 | -2
# 1 | -0.9
# 2 | 0.2 
# and so on

x=np.linspace(0,100)
y=m*x+b
```

```python
#
plt.figure(figsize=(12,12))
plt.plot(x,y)
plt.show()
```

you can add labels and make the graph super. But we will move on to a parabola

```python
# declare some variables -
#DENNIS CHANGES #BUT KEITH CHANGED

m1=0.5
m2=-1.3
b0=5
x=np.linspace(0,100)
y=m2*x**2+m1*x+b0
```

```python
plt.figure(figsize=(12,12))
plt.plot(x,y)
plt.show()
```

This is a nice example of an upside down parabola, but go ahead and change the parameters ($m_{2},m_{1},b_{0}$) to make other examples if you like.

Notice that in the case of $m_{2}=-1.3 ,m_{1}=0.5 ,b_{0}=5$ that as x gets larger ($x\rightarrow100$) $y$ became huge and negative. 

We can plot a lot of functions with this method. In physics the typical $\frac{1}{x^{2}}$ behavior is used in many applications. We look at this next.

```python
# declare some variables
mA=10.1
x=np.linspace(1,100)
y=mA*x**(-2)
plt.figure(figsize=(12,12))
plt.plot(x,y)
plt.show()
```

We made $y=\frac{m_{A}}{x^{2}}$

The astrophysics community calls $y$ a steeply falling function of the variable $x$ for this reason.
It is actually useful to plot it after taking the logarithm of both sides. There are two ways of doing this. The first is to ask the plotter to just change the scale of the plot to a log-scale on both axes

```python
logx=np.log10(x)
logy=np.log10(y)
plt.figure(figsize=(12,12))
plt.subplot(211)
plt.title("Logscale")
plt.plot(x,y)
plt.xscale('log')
plt.yscale('log')
plt.subplot(212)
plt.title("Actually taking the log10")
plt.plot(logx,logy)
plt.show()
```

Notice the difference in the x axis tick marks. In the logscale plot, the ticks are not equidistant. But, when taking the logarithm in the second plot, the x axis labels are for the variable $log(x)$ and so they are equidistant

We can find the slope of the line with a couple points next. You can also chose which points to use in the line after.

```python
slope=(logy[1]-logy[0])/(logx[1]-logx[0])
print(slope)
```

```python
element1=10
element2=5
slope=(logy[element1]-logy[element2])/(logx[element1]-logx[element2])
print(slope)
```

So the slope is -2. This is the power of the variable x that we took the logarithm of and the slope of this line shows the property of logarithms: $log(x^{p})=p*log(x)$
